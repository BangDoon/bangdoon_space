```python
import cv2
from matplotlib import pyplot as plt
```

```python
#영상 읽어서 그레이 스케일로 변환
img = cv2.imread('./dataset/train/CASE31/image/CASE31_01.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 8x8 크기로 축소 ---①
gray = cv2.resize(gray, (150,150))
# 영상의 평균값 구하기 ---②
avg = gray.mean()
# 평균값을 기준으로 0과 1로 변환 ---③
bin = 1 * (gray > avg)
print(bin)

result = 0

for i in bin.reshape(-1,) :
    if(int(i)==0):
        result+=1

print(result)
plt.imshow(img)
```
