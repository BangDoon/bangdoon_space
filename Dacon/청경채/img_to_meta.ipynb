```python
import cv2
from matplotlib import pyplot as plt
```

```python
#    
img = cv2.imread('./dataset/train/CASE31/image/CASE31_01.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 8x8   ---
gray = cv2.resize(gray, (150,150))
#    ---
avg = gray.mean()
#   0 1  ---
bin = 1 * (gray > avg)
print(bin)

result = 0

for i in bin.reshape(-1,) :
    if(int(i)==0):
        result+=1

print(result)
plt.imshow(img)
```
